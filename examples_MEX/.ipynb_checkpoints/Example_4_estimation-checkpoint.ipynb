# Example 4: Estimating the Growth Factors

The previous examples use the growth factors that were estimated for the national-level report. This example shows how to use the function 'estimate' from the 'calibration.py' in order to obtain the growth factors of the development indicators.

## Import the necessary libraries

```python
import warnings
warnings.filterwarnings('ignore')
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
import ppi
import calibration as cal
```

## Load data


```python
data = pd.read_csv('final_sample_normalized.csv', sep='\t', header=0, low_memory=False, encoding='utf-16') # indicators
A = np.loadtxt('network.csv') # network
phi, tau = np.loadtxt('governance_params.csv') # governance parameters
```

## Prepare data for PPI

```python
series = data[[str(year) for year in range(2006, 2017)]].values # extract the time series from the data
I0, T = ppi.get_targets(series) # use the helper function get_targets() to transform the final values into targets
R = data['instrumental'].values # extract the variable indicating which indicators are instrumental
colors = dict(zip(data.index, data.goalColor)) # build a dictionary mapping each indicator to its SDG color
```

## Estimate growth factors

To estimate the vector of growth factors, it is necessary to provide the number of periods that under which the model should converge (the variable name is 'steps'). It is also necessary to provide the number of independent Monte Carlo simulations that will be performed every time a growth factor is changed (the variable name is 'sample_size'). In addition, an error tolerance with respect to 'steps' should be specified (the variable name is 'dev_lim'). The estimation procedure will search for the vector of growth factors until the mean average convergence time of all indicators is close enough to 'steps' under the provided error threshold. A larger 'sample_size' allows more precision and, consequently, can deal with a smaller 'dev_lim'. This however, comes with a computational cost. This example is rather illustrative, so it assumes a large error threshold (of 4 periods) and 5 Monte Carlo simulations.

Finally, the script will run in parallel on 4 independent processes. Most modern computers have multiple cores, so they can handle a 4 parallel proceses. However, should the user have a more powerful machine, she can adjust the 'parallel_processes' parameter to fully exploit parallelization.

```python
%%time
# the line above prints the computation time after the estimation has finished
steps = 30 # the number of periods under which the model should converge
sample_size = 5 # the number of Monte Carlo simulations
dev_lim = 4 # the convergence error threshold
parallel_processes = 4 # the number of processes to be run in parallel


alphas, volatility = cal.estimate(I0, T, A=A, R=R, phi=phi, tau=tau, 
                                 steps=steps, sample_size=sample_size,
                                 dev_lim=dev_lim, parallel_processes=parallel_processes)
```

## Plot the estimated growth factors and their relation to other variables

```python
fig = plt.figure(figsize=(14,3))
plt.subplot(1,3,1)
for i in range(len(A)):
    plt.plot(T[i], alphas[i], '.', mfc=colors[i], mec='white', markersize=14)
plt.xticks([])
plt.xlabel('goal/target', fontsize=14)
plt.ylabel('growth factor', fontsize=14)

plt.subplot(1,3,2)
for i in range(len(A)):
    plt.plot(T[i]-I0[i], alphas[i], '.', mfc=colors[i], mec='white', markersize=14)
plt.xticks([])
plt.xlabel('target-indicator gap', fontsize=14)
plt.ylabel('growth factor', fontsize=14)

plt.subplot(1,3,3)
for i in range(len(A)):
    plt.plot(A.sum(axis=0)[i], alphas[i], '.', mfc=colors[i], mec='white', markersize=14)
plt.xticks([])
plt.xlabel('net incoming spillovers', fontsize=14)
plt.ylabel('growth factor', fontsize=14)

plt.tight_layout()
plt.show()
```

## Plot these growth factors against the ones from the national-level report

Since this example allows large error thresholds and low precision through few Monte Carlo simulations, one should expect discrepancies between both estimations. Nevertheless, robustness of the estimation heuristic should be demonstrated through a positive correlation between them.

```python
alphas_rep = np.loadtxt('alphas.csv') # growth factors
for i in range(len(A)):
    plt.plot(alphas_rep[i], alphas[i], '.', mfc=colors[i], mec='white', markersize=14)
plt.xticks([])
plt.xlabel('growth factor from report', fontsize=14)
plt.ylabel('estimated growth factor', fontsize=14)

plt.tight_layout()
plt.show()

print('Pearson correlation =', np.round(st.pearsonr(alphas_rep, alphas)[0],2))
```
